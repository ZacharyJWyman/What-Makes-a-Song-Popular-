```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv(r"D:\Data Science\Spotify API\data\data.csv")
```

```python
fig, ax = plt.subplots(6,1,figsize=(12,32))

cols = ['acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'loudness']
data = data[:1000]
i = 0
title = ['test1', 'test2', 'test3', 'test4', 'test5', 'test6']
for col in cols:
    sns.scatterplot(ax = ax[i], x = col, y = 'popularity', data = data).set_title(title[i])
    i+=1
    
```
